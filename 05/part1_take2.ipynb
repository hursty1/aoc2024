```python
rules = []
data = []
```

```python
data_=False
with open('sample.txt') as f:
    d = f.readlines()
    for line in d:
        line = line.replace('\n','')
        # print(line)
        if data_:
            data.append(line.split(','))
        else:
            if line == '':
                data_ = True
            else:
                rules.append(line.split('|'))
```

```python
for i,v in enumerate(data[0]):
    print(f'{i}: for {v}')
```

```python
def follow_rules(update):
    idx = {}
    for i, val in enumerate(update): #converts row into list
        idx[val] = i
    for a, b in rules:
        print(idx)
        print(a, '', b)
        if a in idx and b in idx and not idx[a] < idx[b]:
            return False, 0
    return True, update[len(update)//2]
```

```python
running_total = 0
```

```python
running_total = 0
for update in data:
    good, mid = follow_rules(update)
    # print(mid)
    if good:
        running_total += int(mid)
running_total
```
