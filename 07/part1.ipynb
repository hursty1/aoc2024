```python
data = {}
with open('data.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        data[int(line.split(':')[0])] = [int(a) for a in line.split(':')[1].replace('\n','').strip().split(' ')]
```

```python
data
```

```python
operators = ['+','*']
```

```python
from itertools import product
```

```python
p =product(['+','*'],repeat=2)
```

```python
r = 0
for item in data:
    vals = data[item]
    for o in product(operators, repeat=len(vals) -1):
        expression = str(vals[0])  # Start with the first value
        for i, op in enumerate(o):
            expression += f" {op} {vals[i + 1]}" 
        rt = eval(expression)
        if rt == item:
            r += rt
        # print(f'item:{item}: eval for {rt}')
r
```

```python
r = 0
for item in data:
    vals = data[item]
    for o in product(operators, repeat=len(vals) -1):
        expression = str(vals[0])
        rr = int(expression)
        for v, o in zip(vals[1:],o):
            rr = eval(f'{rr}{o}{v}')
        # print(rr)
        # rt = eval(expression)
        # print(f'item:{item}: eval for {rr}')
        if rr == item:
            r += rr
            break
r
```
