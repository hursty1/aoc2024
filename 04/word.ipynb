```python
def gridsearch(x, y, direction, char_int):
    #check if not a match at current position
    if grid[y][x] != search_phrase[char_int]:
        return False
    #check if last letter
    if char_int == len(search_phrase) - 1:
        # Word found
        print(f'Found "{search_phrase}" ending at x:{x}, y:{y}')
        return True
    # next pos
    nx = x + direction[0]
    ny = y + direction[1]
    # Check if next position is within bounds
    if 0 <= nx < len(grid[0]) and 0 <= ny < len(grid):
        return gridsearch(nx, ny, direction, char_int + 1)
    else:
        return False
```

```python
directions = [
    (0, 1),  # Down
    (1, 0),  # Right
    (1, 1),  # Down-Right
    (-1, 0), # Left
    (0, -1), # Up
    (-1, -1),# Up-Left
    (1, -1), # Up-Right
    (-1, 1), # Down-Left
]
```

```python
for y in range(0,len(grid)):
    # print(line)
    for x in range(0,len(grid[0])):
        for direction in directions:
            if gridsearch(x,y,direction,0):
                count += 1
        
```
